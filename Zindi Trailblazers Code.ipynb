Then, we need to do some magic for each pollutant. Let's merge the train features and the test features before doing feature engineering, though

```python
# drop the place_id x date column in the test data
test_features = test_data.drop(['Place_ID X Date'], axis=1)


# Now, let's merge the feature sets
features = pd.concat([train_features, test_features])
#Turn the date column into datetime object
features['Date'] = pd.to_datetime(features['Date'])



print(features['Date'].dtype)

# Now we can engineer the features as we wish.
# The dates are not so useful, so let's extract the months and drop the exact dates(They all have the same year, so it is irrelevant)
features['Month'] = features['Date'].dt.month
display(features)



```

Okay, let's separate the train data from the test(We have performed a new feature engineering.) And also seize this opportinity to split the train to train, test and validation sets.

```python
train = features[features.Date.isin(train_features.Date)]
X_train, X_test, y_train, y_test = train_test_split(train, train_target.target, train_size=0.8)
X_train.drop('Date', axis=1)
X_test.drop('Date', axis=1)





# We will need to scale this thing o   
                          
test = features[features.Date.isin(test_features.Date)]
test.drop("Date", axis=1)
```

Since this is a regression problem, Lasso is a first place to start.

```python
lasso = Lasso()

abc = cross_val_score(lasso, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
print(abc.best_score_)
print(abc.score(X_test, y_test))
```
